```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
# Prevent the error "session not created: probably user data directory is already in use"
options.add_argument('--user-data-dir=/tmp/user-data')

# Add a small delay before initializing the webdriver
time.sleep(2)

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)


url = "https://www.flipkart.com/search?q=remote+control+car&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&as-pos=1&as-type=HISTORY&page=1"
driver.get(url)

car_names =[]
car_prices =[]
car_ratings =[]
car_review = []
num_pages = 30

wait = WebDriverWait(driver, 10)

for page in range(num_pages):
  time.sleep(5) # Reduced sleep time

  content = driver.page_source
  soup = BeautifulSoup(content , 'html.parser')

  # Updated selectors based on the website structure
  car = soup.find_all('div',class_ ="slAVV4")
  for i in car :
    names = i.find('a',class_="wjcEIp")
    if names:
      car_names.append(names.get_text(strip = True))
    else :
      car_names.append('N/A')

    prices = i.find('div',class_="Nx9bqj")
    if prices:
      car_prices.append(prices.get_text(strip=True))
    else:
      car_prices.append("N/A")

    ratings = i.find('div',class_="XQDdHH")
    if ratings:
      car_ratings.append(ratings.get_text(strip=True))
    else:
      car_ratings.append("N/A")

    review= i.find('span',class_ = "Wphh3N")
    if review:
      car_review.append(review.get_text(strip=True))
    else:
      car_review.append("N/A")
  try:
    # Updated XPath for next button based on the website structure
    nxt_button = wait.until(EC.element_to_be_clickable((By.XPATH,'//a[@class="_9QVEpD"]')))
    nxt_button.click()
  except Exception as e:
    print(e)
    break
driver.quit()

data = {"car_names ":car_names,
        "car_prices": car_prices,
        "car_ratings": car_ratings,
        "car_review":car_review}
df = pd.DataFrame(data)
print(df)
df.to_csv("Remote_Car.csv")
```
